```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 1000) #for some reason not producing the right range
```

```python
sint = np.sin(t)
```

```python
cost = np.cos(t)
```

```python

plt.figure(figsize = (6,12))
plt.plot(sint)
plt.plot(cost)
plt.xlabel("Time [hr]")
plt.ylabel("Wave Height [m]")
plt.title("Wave Heights at Fraggle Rock Cove")
#plt.legend(location = "best")
```
